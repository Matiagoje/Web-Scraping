```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.scrapethissite.com/pages/forms/?page_num='
data = []
 
for page in range(1,5):
    response = requests.get(url + str(page))
    html_page = BeautifulSoup(response.content, 'html.parser')
    html_page = BeautifulSoup(html_page.prettify(), 'html.parser')

    table = html_page.find(class_ = 'table')
    table_rows = table.find_all(class_ = 'team')
    for column in table_rows:
        rows = column.find_all('td')
        row_data = [content.text.strip() for content in rows]
        data.append(row_data)
```

```python
table_headers=  table.find_all('th')
table_headers
```

```python
headers = [header.text.strip() for header in table_headers]
print(headers)
```

```python
df = pd.DataFrame(data, columns=headers)
df
```
